```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("Datasets/Titanic_Dataset.csv")
df.head()
```

## 1) Categorical Data 

a) Countplot

```python
# sns.countplot(x='Survived', data=df)
# df['Survived'].value_counts().plot(kind='bar')

sns.countplot(x='Pclass', data=df)
```

b) Piechart

```python
# If you want the same information in percentage then use piechart
df['Survived'].value_counts().plot(kind='pie', autopct= '%.2f')
# autopct is used to calculate the percentage 
```

## 2) Numerical Data

### a) Histogram

```python
import matplotlib.pyplot as plt

plt.hist(df['Age'], bins=20)
```

### b) Distplot

```python
sns.distplot(df['Age'])  
# => But distplot() is removed from the seborn
```

```python
# => Alternate way to do that
sns.histplot(df['Age'], bins=20, kde=True, color='skyblue') 
```

```python
# Or eles if you only want the KDE(Kernal Density Estimation) 
sns.kdeplot(df['Age'], fill=True, color='skyblue')
```

### c) Boxplot

```python
sns.boxplot(x='Fare', data=df)

# There are lot of outliers in the FARE, For one sit someone pays $500
```

```python
sns.boxplot(x='Age', data=df)
# They generally used where there is some noise in the dataset
```

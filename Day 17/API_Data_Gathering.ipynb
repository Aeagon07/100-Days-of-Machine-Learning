```python
import pandas as pd
import requests

emp_df = pd.DataFrame()

for i in range(1, 400):
    resp = requests.get('https://api.themoviedb.org/3/movie/top_rated?api_key=93e55eceb81589620395dfa04f96c1f7&language=en-US&page={}'.format(i))
    # print(resp.status_code)
    temp_df = pd.DataFrame(resp.json()['results'])[['id', 'original_language', 'title', 'overview', 'popularity', 'vote_average', 'vote_count']]
    emp_df = pd.concat([emp_df, temp_df], ignore_index= True) # ignore_index = True -> which create own indexing
    
```

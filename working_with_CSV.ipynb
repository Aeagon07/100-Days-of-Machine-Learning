 3. From the Random URL, we can also open CSV files.

```python
import requests
from io import StringIO

url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firstfox/66.0"}
req = requests.get(url, headers=headers)
data = StringIO(req.text)

pd.read_csv(data)
```

14. Convetors

```python
def rename(name):
    if name == "Royal Challenges Banglore":
        return "RCB"
    else:
        return name
```

```python
rename("Royal Challenges Banglore")
```

```python
pd.read_csv("somedta.csv", converters={'team1': rename})
# Where team1 is column name and it converted them to RCB
```
